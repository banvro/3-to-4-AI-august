```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['cgpa'], data['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(1, 20):
    km = KMeans(n_clusters = i)
    km.fit_predict(data)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 20), wcss)
```

```python
km = KMeans(n_clusters = 4)
```

```python
xyz = data.iloc[:, :].values
```

```python
xyz
```

```python
y_pred = km.fit_predict(xyz)
```

```python
xyz
```

```python
xyz[y_pred == 3]
```

```python
plt.scatter(xyz[y_pred == 0, 0], xyz[y_pred == 0, 1], color = "red")
plt.scatter(xyz[y_pred == 1, 0], xyz[y_pred == 1, 1], color = "green")
plt.scatter(xyz[y_pred == 2, 0], xyz[y_pred == 2, 1], color = "yellow")
plt.scatter(xyz[y_pred == 3, 0], xyz[y_pred == 3, 1], color = "blue")
```
